```python
#q1
# Initial list
L = [10, 20, 30, 40, 50, 60, 70, 80]

# i. Add 200 and 300 to L using append() and extend()
L.append(200)  # Adding 200
L.extend([300])  # Adding 300 using extend()

# ii. Remove 10 and 30 from L using remove()
try:
    L.remove(10)  # Removing 10
    L.remove(30)  # Removing 30
except ValueError:
    print("Value not found in the list for removal.")

# iii. Sort L in ascending order using sort()
L.sort()  # Sorting in ascending order

# iv. Sort L in descending order using sort(reverse=True)
L.sort(reverse=True)  # Sorting in descending order

# Output  final list
print(L)

#q2
# Tuple of marks scored
scores = (45, 89.5, 76, 45.4, 89, 92, 58, 45)

# i. Identify the highest score and its index in the tuple
highest_score = max(scores)
highest_index = scores.index(highest_score)
print(f"The highest score is {highest_score} and its index is {highest_index}")

# ii. Find the lowest score and count how many times it appears
lowest_score = min(scores)
lowest_count = scores.count(lowest_score)
print(f"The lowest score is {lowest_score} and it appears {lowest_count} times")

# iii. Reverse the tuple and return it as a list
reversed_scores = list(scores[::-1])
print("Reversed tuple as a list:", reversed_scores)

# iv. Check if a specific score (76) input by the user is present in the tuple
user_score = float(input("Enter a score to check if it is in the tuple: "))
if user_score in scores:
    index_of_score = scores.index(user_score)
    print(f"The score {user_score} is present at index {index_of_score}")
else:
    print(f"The score {user_score} is not present in the tuple.")

#q3
import random
import math

# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

# Create a list of 100 random numbers between 100 and 900
random_numbers = [random.randint(100, 900) for _ in range(100)]

# i. All odd numbers using list comprehension
odd_numbers = [num for num in random_numbers if num % 2 != 0]

# ii. All even numbers using list comprehension
even_numbers = [num for num in random_numbers if num % 2 == 0]

# iii. All prime numbers using list comprehension
prime_numbers = [num for num in random_numbers if is_prime(num)]

# Print the results
print(f"Odd numbers: {odd_numbers}")
print(f"Even numbers: {even_numbers}")
print(f"Prime numbers: {prime_numbers}")

#q4
# Define sets A and B
A = {34, 56, 78, 90}
B = {78, 45, 90, 23}

# i. Find the unique scores achieved by both teams (union)
unique_scores = A | B  # Union using | operator
print("Unique scores (Union):", unique_scores)

# ii. Identify the scores that are common to both teams (intersection)
common_scores = A & B  # Intersection using & operator
print("Common scores (Intersection):", common_scores)

# iii. Find the scores that are exclusive to each team (symmetric difference)
exclusive_scores = A ^ B  # Symmetric difference using ^ operator
print("Exclusive scores (Symmetric Difference):", exclusive_scores)

# iv. Check if the scores of team A are a subset of team B, and if team B's scores are a superset of team A
is_A_subset_B = A.issubset(B)  # Checks if A is a subset of B
is_B_superset_A = B.issuperset(A)  # Checks if B is a superset of A
print(f"Is A a subset of B? {is_A_subset_B}")
print(f"Is B a superset of A? {is_B_superset_A}")

# v. Remove a specific score X (input by the user) from set A if it exists
score_to_remove = int(input("Enter a score to remove from team A: "))

#q5
# Initial dictionary
data = {"city": "New York", "population": 8419600, "area": 468.9}

# i. Rename the key 'city' to 'location'
data["location"] = data.pop("city")  # pop removes 'city' and returns its value, which we assign to 'location'

# ii. Add a new key-value pair
data["mayor"] = "Bill de Blasio"  # Adding a new key-value pair for mayor

# iii. Remove a key-value pair (e.g., remove 'area')
data.pop("area")  # Removes the key 'area
```
